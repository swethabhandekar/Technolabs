```python
def isClosed(row):
    if row['status'] == 'closed' or row['status'] == 'acquired':
        return 1
    else:
        return 0
df['isClose'] = df.apply(lambda rw: isClosed(rw),axis=1)
```

```python
founded_date = pd.to_datetime(df['founded_at'],format="%Y-%m-%d")

```

```python
founded_date
```

```python
def closed_at(row):
    if row['status'] == 'operating' or row['status'] == 'ipo':
        return '2013-01-01' 
    else:
        return row['closed_at']
df['closed_at'] = df.apply(lambda rw: closed_at(rw),axis=1)
df['closed_at'] = df['closed_at'].fillna('2013-01-01')

```

```python
closed_date = pd.to_datetime(df['closed_at'],format="%Y-%m-%d")
        
```

```python
closed_date
```

```python
df['activeDays'] = closed_date-founded_date
df['activeDays'] = df['activeDays'].dt.days

```
